```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
path_image="../output_video/cropped_image.jpg"

image=cv2.imread(path_image)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

#take the top of the imageµ

```python
top_half=image[0:int(image.shape[0]/2),:]
plt.imshow(top_half)
```

Cluster the image into two clustersµ

```python
image_2d=top_half.reshape(-1,3)

#kmean cluster
kmean=KMeans(n_clusters=2,random_state=0)
kmean.fit(image_2d)


#label
labels=kmean.labels_

#reshape label
clustered_image=labels.reshape(top_half.shape[0],top_half.shape[1])



plt.imshow(clustered_image)
```

```python
 corner_clusters=[clustered_image[0,0],clustered_image[0,-1],clustered_image[-1,0],clustered_image[-1,-1]]
 non_player_cluster=max(set(corner_clusters),key=corner_clusters.count)
 print(non_player_cluster)
```

```python
player_cluster=1-non_player_cluster
print(player_cluster)
```

```python
kmean.cluster_centers_[player_cluster]
```
